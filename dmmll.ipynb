```
dataset=pd.read_csv('titanic.csv',usecols=["Age","Cabin","Embarked"])
dataset
```

```
dataset.isnull().sum()
```

```
mean = dataset['Age'].mean()
mean
```

```
def input_mean(dataset,mean):
  dataset['Age_mean']=dataset['Age'].fillna(mean)
  return dataset
```

```
input_mean(dataset,mean)
```

```
dataset['Age'].plot(kind="kde")
dataset['Age_mean'].plot(kind="kde")
```

```
median = dataset['Age'].median()
median
```

```
def input_median(dataset,median):
  dataset['Age_median']=dataset['Age'].fillna(median)
  return dataset
```

```
input_median(dataset,median)
```

```
dataset['Age'].plot(kind="kde")
dataset['Age_mean'].plot(kind="kde")
dataset['Age_median'].plot(kind="kde")
```

```
Mode = dataset['Age'].mode()[0]
Mode
```

```
def input_mode(dataset,Mode):
  dataset['Age_mode']=dataset['Age'].fillna(Mode)
  return dataset
```

```
input_mode(dataset,Mode)
```

```
dataset['Age'].plot(kind="kde")
dataset['Age_mean'].plot(kind="kde")
dataset['Age_median'].plot(kind="kde")
dataset['Age_mode'].plot(kind="kde")
```
